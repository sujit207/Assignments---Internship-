# Problen-1- Scrape the dtls of most viewed videos on YouTube from Wikipedia

```python
driver= webdriver.Chrome(r"C:\Users\arkay\Downloads\chromedriver_win32 (1)\chromedriver.exe")
url='https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
driver.get(url)
rank=[]
name=[]
artist=[]
date=[]
views=[]
r_tags=driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[1]/tbody/tr/td[1]")
for i in r_tags:
    dataset=i.text
    rank.append(dataset)

n_tags=driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[1]/tbody/tr/td[2]")
for i in n_tags:
    dataset=i.text
    name.append(dataset)
    
a_tags=driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[1]/tbody/tr/td[3]")
for i in a_tags:
    dataset=i.text
    artist.append(dataset)

d_tags=driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[1]/tbody/tr/td[5]")
for i in d_tags:
    dataset=i.text
    date.append(dataset)

v_tags=driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[1]/tbody/tr/td[4]")
for i in v_tags:
    dataset=i.text
    views.append(dataset)
```

```python
Youtube=pd.DataFrame({})
Youtube['Rank']=rank
Youtube['Artist']=artist
Youtube['Name']=name
Youtube['Date']=date
Youtube['Views_In_Billions']=views
Youtube
```

# Probelm 4- Scrape the details of State-wise GDP of India from statisticstime.com.

```python
driver= webdriver.Chrome(r"C:\Users\arkay\Downloads\chromedriver_win32 (1)\chromedriver.exe")
url='http://statisticstimes.com/economy/gdp-of-indian-states.php'
driver.get(url)
```

```python
rank=[]
state=[]
gsdp1=[]
gsdp2=[]
share=[]
gdp=[]

r_tags=driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr/td[1]")
for i in r_tags:
    dataset=i.text
    rank.append(dataset)

s_tags=driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr/td[2]")
for i in s_tags:
    dataset=i.text
    state.append(dataset)

g1_tags=driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr/td[3]")
for i in g1_tags:
    dataset=i.text
    gsdp1.append(dataset)

g2_tags=driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr/td[4]")
for i in g2_tags:
    dataset=i.text
    gsdp2.append(dataset)

sh_tags=driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr/td[5]")
for i in sh_tags:
    dataset=i.text
    share.append(dataset)

gdp_tags=driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr/td[6]")
for i in gdp_tags:
    dataset=i.text
    gdp.append(dataset)

GDP_Indian_States=pd.DataFrame({})
GDP_Indian_States['Rank']=rank
GDP_Indian_States['State_Name']=state
GDP_Indian_States['GSDP(18-19)']=gsdp1
GDP_Indian_States['GSDP(17-18)']=gsdp2
GDP_Indian_States['Share(2017)']=share
GDP_Indian_States['GDP($ Billion)']=gdp
GDP_Indian_States
```

# Problem 6- Scrape the details of top 100 songs on billiboard.com.

```python
driver= webdriver.Chrome(r"C:\Users\arkay\Downloads\chromedriver_win32 (1)\chromedriver.exe")
url='https://www.billboard.com/charts/hot-100'
driver.get(url)

Song=[]
Artist=[]
Last_Week_Rank=[]
Peak_Rank=[]
Weeks_On_Board=[]

# lets scrap the data from the server
song=driver.find_elements_by_xpath("//span[@class='chart-element__information__song text--truncate color--primary']")
for i in song:
    so_song=i.text
    Song.append(so_song)
    S=pd.DataFrame(Song)
    S=S.rename({0: 'Song'},axis=1)
artist=driver.find_elements_by_xpath("//span[@class='chart-element__information__artist text--truncate color--secondary']")
for i in artist:
    ar_song=i.text
    Artist.append(ar_song)
    A=pd.DataFrame(Artist)
    A=A.rename({0: 'Artist'},axis=1)
lwr=driver.find_elements_by_xpath("//div[@class='chart-element__metas chart-element__metas--large display--flex flex--y-center']/div[2]")
for i in lwr:
    lw_song=i.text
    Last_Week_Rank.append(lw_song)
    L=pd.DataFrame(Last_Week_Rank)
    L=L.rename({0: 'Last_Week_Rank'},axis=1)
pr=driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--peak']")
for i in pr:
    pr_song=i.text
    Peak_Rank.append(pr_song)
    P=pd.DataFrame(Peak_Rank)
    P=P.rename({0: 'Peak_Rank'},axis=1)   
wob=driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--week']")
for i in wob:
    w_song=i.text
    Weeks_On_Board.append(w_song)
    W=pd.DataFrame(Weeks_On_Board)
    W=W.rename({0: 'Weeks_On_Board'},axis=1)   
#Lets create a table of top 100 movies data
movies=pd.concat([S,A,L,P,W] , axis=1)
movies
```

# Problem 8 -Scrape the details of Highest selling novels.

```python
driver= webdriver.Chrome(r"C:\Users\arkay\Downloads\chromedriver_win32 (1)\chromedriver.exe")
url='https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare'
driver.get(url)

book_name = []
author_name  = []
volumes_sold  = []
publisher  = []
genre  = []

Task1 = driver.find_elements_by_xpath("//tbody/tr/td[2]")

for qt in Task1:
    qx = qt.text
    book_name.append(qx)

book_name1 =pd.DataFrame(book_name)
book_name2= book_name1.rename(columns = { 0: 'Book Name'})

Task2 = driver.find_elements_by_xpath("//tbody/tr/td[3]")

for qm in Task2:
    ql = qm.text
    author_name.append(ql)

author_name1 = pd.DataFrame(author_name)
author_name2 = author_name1.rename(columns= {0:'Author Name'})

Task3 = driver.find_elements_by_xpath("//tbody/tr/td[4]")

for wp in Task3:
    wm = wp.text
    volumes_sold.append(wm)
volumes_sold[0:5]

volumes_sold1 = pd.DataFrame(volumes_sold)
volumes_sold2 =  volumes_sold1.rename (columns ={ 0: 'Volumes Sold'})

Task4 = driver.find_elements_by_xpath("//tbody/tr/td[5]")

for ap in Task4:
    ac = ap.text
    publisher.append(ac)

publisher1 = pd.DataFrame(publisher)
publisher2 = publisher1.rename(columns = { 0 : 'Publisher'})

Task5 = driver.find_elements_by_xpath("//tbody/tr/td[6]")

for vp in Task5:
    qf = vp.text
    genre.append(qf)

genre1 = pd.DataFrame(genre)
genre2 = genre1.rename (columns = { 0: 'Genre'})

Highest_Selling_Novels = pd.concat( [book_name2 , author_name2 , volumes_sold2 , publisher2 ,genre2]  , axis = 1)
Highest_Selling_Novels 
```

# Problem 10- Details of Datasets from UCI machine learning repositories.

```python
driver= webdriver.Chrome(r"C:\Users\arkay\Downloads\chromedriver_win32 (1)\chromedriver.exe")
url='https://archive.ics.uci.edu/'
driver.get(url)
```

```python
tag=driver.find_element_by_xpath("/html/body/table[2]/tbody/tr/td/span/b/a")
drp=tag.get_attribute("href")
driver.get(drp)

name=[]
datatype=[]
task=[]
attributetype=[]
instances=[]
attribute=[]
year=[]

name_tags=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[1]/table/tbody/tr/td[2]/p/b/a")
for i in name_tags:
    dataset=i.text
    name.append(dataset)
d_tags=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[2]/p")
for i in d_tags:
    dataset=i.text
    datatype.append(dataset)
t_tags=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[3]/p")
for i in t_tags:
    dataset=i.text
    task.append(dataset)
a_tags=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[4]/p")
for i in a_tags:
    dataset=i.text
    attributetype.append(dataset)
i_tags=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[5]/p")
for i in i_tags:
    dataset=i.text
    instances.append(dataset)
na_tags=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[6]/p")
for i in na_tags:
    dataset=i.text
    attribute.append(dataset)
y_tags=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[7]/p")
for i in y_tags:
    dataset=i.text
    year.append(dataset)

UCI_dataset=pd.DataFrame({})
UCI_dataset['Datatype']=datatype
UCI_dataset['Deafult_task']=task
UCI_dataset['Attribute']=attributetype
UCI_dataset['Instances']=instances
UCI_dataset['No_of_Attribure']=attribute
UCI_dataset['Year']=year

UCI_dataset
```
