# Problem 2

```python
p=requests.get('https://www.imdb.com/search/title/?count=100&groups=top_1000&sort=user_rating')
p.content
```

```python
soup=BeautifulSoup(p.content,'html.parser')
year_list=[]
rating_list=[]
movie_name=[]
```

```python
year_tags=soup.find_all('span',class_="lister-item-year text-muted unbold")
year_tags[0:4]
```

```python
for i in year_tags:
    year_list.append(i.get_text())
year_list[0:4]
```

```python
rating_tags=soup.find_all('div',class_="inline-block ratings-imdb-rating")
rating_tags[0:4]
```

```python
for i in rating_tags:
    rating=""
    for j in i.find_all('strong'):
        rating=rating+" "+j.get_text()
        rating_list.append(rating)
rating_list[0:4]
```

```python
movie_tags=soup.find_all('h3',class_="lister-item-header")
movie_tags[0:4]
```

```python
for i in movie_tags:
    movie=""
    for j in i.find_all('a'):
        movie=movie+" "+j.get_text()
        movie_name.append(movie)
movie_name[0:4]
```

```python
print(len(movie_name), len(year_list), len(rating_list))
```

```python
imdb=pd.DataFrame({})
imdb['movie']=movie_name
imdb['year_of_release']=year_list
imdb['rating']=rating_list
imdb
```
