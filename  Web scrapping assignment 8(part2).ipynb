# Prob 5 Scrape the details of trending repositories on Github.com.

```python
driver= webdriver.Chrome(r"C:\Users\arkay\Downloads\chromedriver_win32 (1)\chromedriver.exe")
url='https://github.com/'
driver.get(url)
time.sleep(5)
```

```python
r_title=[]
r_desc=[]
c_count=[]
l_used=[]
```

```python
r_tags=driver.find_elements_by_xpath("/html/body/div[4]/main/div[3]/div/div[2]/article/h1/a/span")
for i in r_tags:
    dataset=i.text
    r_title.append(dataset)
rd_tags=driver.find_elements_by_xpath("/html/body/div[4]/main/div[3]/div/div[2]/article/p")
for i in rd_tags:
    dataset=i.text
    r_desc.append(dataset)
            
c_tags=driver.find_elements_by_xpath("/html/body/div[4]/main/div[3]/div/div[2]/article/div[2]/a[2]")
for i in c_tags:
    dataset=i.text
    c_count.append(dataset)
l_tags=driver.find_elements_by_xpath("/html/body/div[4]/main/div[3]/div/div[2]/article/div[2]/span[1]/span[2]")
for i in l_tags:
    dataset=i.text
    l_used.append(dataset)

GH=pd.DataFrame({})
GH['Repository_Name']=r_title
GH['Contributors_count']=c_count
GH['Language_Used']=l_used
GH
```

# Prob -7 Scrape the details of Data science recruiters from naukri.com.


```python
driver= webdriver.Chrome(r"C:\Users\arkay\Downloads\chromedriver_win32 (1)\chromedriver.exe")
url='https://www.naukri.com/data-science-recruiters'
driver.get(url)
time.sleep(5)
driver.maximize_window()
```

```python
driver.find_element_by_xpath("/html/body/div[1]/div/div/ul[1]/li[2]/a").click()
time.sleep(5)
```

```python
name=[]
des=[]
com=[]
skil=[]
loc=[]
```

```python
n_tags=driver.find_elements_by_xpath("//*[@id='tabP-1']/div/div/div[1]/div[1]/p/a[1]/span")
for i in n_tags:
    dataset=i.text
    name.append(dataset)
d_tags=driver.find_elements_by_xpath("//*[@id='tabP-1']/div/div/div[1]/div[1]/p/span[1]")
for i in d_tags:
    dataset=i.text
    des.append(dataset)
c_tags=driver.find_elements_by_xpath("//*[@id='tabP-1']/div/div/div[1]/div[1]/p/a[2]")
for i in c_tags:
    dataset=i.text
    com.append(dataset)
s_tags=driver.find_elements_by_xpath("//*[@id='tabP-1']/div/div/div[1]/div[2]")
for i in s_tags:
    dataset=i.text
    skil.append(dataset)
l_tags=driver.find_elements_by_xpath("//*[@id='tabP-1']/div/div/div[1]/div[1]/p/span[2]/small")
for i in l_tags:
    dataset=i.text
    loc.append(dataset)
```

```python
N=pd.DataFrame({})
N['Name']=name
N['Designation']=des
N['Company']=com
N['Skills_they_hire_for']=skil
N
```
